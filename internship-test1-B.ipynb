```python
#----------Running the sum-------------
class Solution:
    def runningSum(self, nums):
        output = []
        total = 0
        for i in nums:
            total += i
            output.append(total)
        return output


s = Solution()
print(s.runningSum([1,2,3,4,5]))
```

```python
#------shuffle the array-----------
class Solution:
    def shuffle(self, nums, n): #-> List[int]:
        output = []
        for i in range(n):
            output.append(nums[i])
            output.append(nums[i+n])
        return output
s = Solution()
print(s.shuffle([1,2,4,5,6,7,9,3],4))
```

```python
#-------another solution for shuffle the array-----------
class Solution:
    def shuffle(self, nums,n):# -> List[int]:
        output = [0]*(2*n)
        for i in range(n):
            output[2*i] = nums[i]
            output[2*i+1] = nums[i+n]
        return output
s = Solution()
print(s.shuffle([1,2,4,5,6,7,9,3],4))
```

```python
#-------------Kids With the Greatest Number of Candies ------------------
class Solution:
    def kidsWithCandies(self, candies, extraCandies): #-> List[bool]:
        output = []
        for i in candies:
            if i + extraCandies >= max(candies):
                output.append(True)
            else :
                output.append(False)
        return output
s = Solution()
print(s.kidsWithCandies([2,3,7,9,4,1],5))
```
